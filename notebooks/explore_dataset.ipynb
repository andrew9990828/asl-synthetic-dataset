```python
import matplotlib.pyplot as plt
from PIL import Image
import os
import random

root = "../asl_abstract_dataset"

letters = os.listdir(root)
letters = [l for l in letters if len(l) == 1]

for _ in range(5):
    letter = random.choice(letters)
    sample = random.choice(os.listdir(os.path.join(root, letter)))

    img = Image.open(os.path.join(root, letter, sample))

    plt.imshow(img)
    plt.title(f"Letter: {letter}")
    plt.axis("off")
    plt.show()
```
